```python
#importando datasets
import pandas as pd

cadastro = pd.read_csv('C:/Users/ricardo/Downloads/TRABALHO_BOOTCAMP_DATA_ENGINEERING/cadastro.csv')
dados_medicos = pd.read_csv('C:/Users/ricardo/Downloads/TRABALHO_BOOTCAMP_DATA_ENGINEERING/dados_medicos.csv')

```

```python
#criando funcao para remover caracteres especiais
def corrigir_nomes(nome):
    nome = nome.replace('.', '').replace('ç', 'c').replace('ô', 'o').replace('á', 'a').replace('é', 'e').replace('í', 'i').replace('ó', 'o').replace('ú', 'u').replace('ã', 'a').replace('â', 'a').replace('õ', 'o')
    return nome
```

```python
#removendo caracteres especiais do nome
cadastro['nome'] = cadastro['nome'].apply(corrigir_nomes)
```

```python
#removendo caracteres especiais do nome da mae
cadastro['mae'] = cadastro['mae'].apply(corrigir_nomes)
```

```python
#verificando datatype do campo de data de nascimento
type(cadastro['data_nasc'])
```

```python
#convertendo campos da tabela cadastro
cadastro['nome'] = cadastro['nome'].astype(str)
cadastro['data_nasc']  = pd.to_datetime(cadastro['data_nasc'])
cadastro['cpf'] = cadastro['cpf'].astype(int)
cadastro['mae'] = cadastro['mae'].astype(str)
cadastro['celular'] = cadastro['celular'].astype(str)
cadastro['email'] = cadastro['email'].astype(str)
cadastro['telefone_fixo'] = cadastro['telefone_fixo'].astype(str)
```

```python
#removendo caracteres especiais do nome
dados_medicos['nome'] = dados_medicos['nome'].apply(corrigir_nomes)
```

```python
#criando funcao para alterar , para . no campo altura
def corrigir_altura(altura):
    altura = altura.replace(',', '.')
    return altura
```

```python
#alterando , por . no campo altura
dados_medicos['altura'] = dados_medicos['altura'].apply(corrigir_altura)
```

```python
#convertendo campos da tabela dados_medicos
dados_medicos['nome'] = dados_medicos['nome'].astype(str)
dados_medicos['data_nasc']  = pd.to_datetime(dados_medicos['data_nasc'])
dados_medicos['cpf'] = dados_medicos['cpf'].astype(int)
dados_medicos['tipo_sanguineo'] = dados_medicos['tipo_sanguineo'].astype(str)
dados_medicos['peso'] = dados_medicos['peso'].astype(int)
dados_medicos['altura'] = dados_medicos['altura'].astype(float)
```
