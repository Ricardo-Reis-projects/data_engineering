```python
#Bar Plots

#importando matplotlib
import matplotlib.pyplot as plt
```

```python
#importando pandas
import pandas as pd
```

```python
#lendo arquivo csv
df = pd.read_csv('top_10_countries.csv')
```

```python
df.rename(columns={'Country / Dependency': 'Country'}, inplace=True)
```

```python
df
```

```python
df_g = df[['Region','Population']]
```

```python
df_g
```

```python
df_g = df_g.groupby(by='Region').sum()
```

```python
df_g
```

```python
#inserindo indice nas linhas no dataframe
df_g.reset_index(inplace=True)
```

```python
df_g
```

```python
# Ordenando o dataframe pelo campo population
df_g.sort_values('Population', ascending=False, inplace=True)
```

```python
df_g
```

```python
# criando o figure and axes do objeto 

# plot area
fig = plt.figure(figsize=(5,5))

#axes
axes = fig.add_axes([0,0,1,1])
axes.set_xlabel('Region')
axes.set_ylabel('Total Population')
axes.set_title('Top 10 Population by Region')

# bar plot

# x => Legenda
# Height => Valor do Grafico
# color => cor das barras
# edgecolor = cor das bordas da barra

plt.bar(x=df_g['Region'], height=df_g['Population'], color=['b','r','y','g'], edgecolor='black')

#pritando o grafico na tela ao inves de retornar no output do jupyter notebook
plt.show()
```

```python
# Horizontal bar plots

# creating the figure and axes / area do grafico

# 1 maneira
#fig = plt.figure(figsize=(5,5))
#axes1 = fig.add_axes([0,0,1,1])
#axes2 = fig.add_axes([1.5,0,1,1])

# 2 maneira
fig,(ax1,ax2)=plt.subplots(nrows=1,ncols=2,figsize=(20,8))

# labels and titles
ax1.set_title('Total Population by Region')
ax1.set_ylabel('Population')
ax1.set_xlabel('Region')


ax2.set_title('Total Population by Country')
ax2.set_ylabel('Population')
ax2.set_xlabel('Country')

#bar plot
ax1.bar(x=df_g['Region'], height=df_g['Population'])

# horizontal bar plot
ax2.barh(y=df['Country'], width=df['Population'])


plt.show()
```
